# Analysis of Sorting Algorithm Performance

```python
import pandas as pd
import numpy as np

import seaborn as sns
figsize = (22, 8)
sns.set_theme(font_scale=1.0, rc={'lines.linewidth': 2.5, 'figure.figsize': figsize})
sns_fig_size = {'height': 8, 'aspect': 2.75}
import matplotlib.pyplot as plt
```

# Test 1
Prior to inlining functions. Compares a functional implementation of selection sort to an imperative one.

```python
results = pd.read_csv('../BenchmarkDotNet.Artifacts/results/FsAlgo.PerfTest.Sorting.Sort-report-1.csv')
results.columns
```

```python
results = results[['Method', 'N', 'Mean [ms]', 'Median [ms]', 'StdDev [ms]', 'Error [ms]']]
results
```

```python
sns.lineplot(data=results, x='N', y='Mean [ms]', hue='Method')
```

```python
sns.lineplot(data=results, x='N', y='StdDev [ms]', hue='Method')
```

# Test 2
Inlining functions, increasing list sizes and comparing to functional merge sort and F# built-in sort.

Note: F# List sort just does a "toArray" before sorting, and so will just contain some overhead.

```python
results = pd.read_csv('../BenchmarkDotNet.Artifacts/results/FsAlgo.PerfTest.Sorting.Sort-report-2.csv')
results = results[['Method', 'N', 'Mean [ms]', 'Median [ms]', 'StdDev [ms]', 'Error [ms]']]
results
```

```python
sns.lineplot(data=results, x='N', y='Mean [ms]', hue='Method')
```

```python
results_subset = results.query('Method.isin(["FunctionalMergeSort", "FSharpSort", "FSharpSortArray"])')
results_subset
```

```python
sns.lineplot(data=results_subset, x='N', y='Mean [ms]', hue='Method')
```

## Comments
Inlining does boost performance a bit. Built-in sort is very efficient as expected. Further tests will need to use bigger lists.

Merge-sort, even in functional form, outdoes more naive approaches, as expected.
